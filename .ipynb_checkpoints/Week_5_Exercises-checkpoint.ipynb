```python
import pandas as pd
import matplotlib.pyplot as plt

#With data aggregation and data visualization answer the following questions:

#What is the change in pct of divorced danes from 2008 to 2020?
df = pd.read_csv('https://api.statbank.dk/v1/data/FOLK1A/CSV?delimiter=Semicolon&CIVILSTAND=TOT%2CF&Tid=2008K1%2C2020K1', sep=';')
print(df)
div_08 = df.iloc[2,2]
tot_08 = df.iloc[0,2]
div_20 = df.loc[3,'INDHOLD']
tot_20 = df.loc[1,'INDHOLD']

pct_div_08 = div_08/tot_08*100
pct_div_20 = div_20/tot_20*100

years = list(set(df.loc[:,'TID']))

plt.bar(years, [pct_div_08, pct_div_20])
plt.title('Percent of people divorced')
plt.xlabel('Years')
plt.ylabel('Percent divorced')
```

```python
#Which of the 5 biggest cities has the highest percentage of 'Never Married'?
df = pd.read_csv('https://api.statbank.dk/v1/data/FOLK1A/CSV?delimiter=Semicolon&CIVILSTAND=TOT%2CU&Tid=2020K3&OMR%C3%85DE=*', sep=';')
#print(df)
#finder total befolkningen i områderne for at sortere listen
areas_total = df.loc[df['CIVILSTAND'] == 'I alt'].loc[:,'OMRÅDE':'INDHOLD']
areas = df.loc[df['CIVILSTAND'] == 'I alt']['OMRÅDE']
#print(areas)
areas_total = areas_total.set_index(areas)

#sortere efter indhold
areas_total = areas_total.sort_values(by='INDHOLD', ascending=False)

#finder de fem største. Der er godt nok regioner med, men oh well
five_biggest = areas_total.iloc[0:5,0]
#print(five_biggest)

unmarried = df.loc[df['CIVILSTAND'] == 'Ugift'].loc[df['OMRÅDE'] == 'København']['INDHOLD']

result = {}
#looper igennem de 5 største områder for at tjekke hvor mange der er ugift i hver af dem
for area in list(five_biggest):
    unmarried = df.loc[df['CIVILSTAND'] == 'Ugift'].loc[df['OMRÅDE'] == area]['INDHOLD']
    total = df.loc[df['CIVILSTAND'] == 'I alt'].loc[df['OMRÅDE'] == area]['INDHOLD']
    unmarried_pct = int(unmarried)/int(total)*100
    result[area] = unmarried_pct
print(result)
#sortere result for at finde det område med flest ugifte
result_sorted = {k: v for k, v in sorted(result.items(), key=lambda item: item[1], reverse=True)}
#printer [0] i listen efter den er sorted
print('Area with highest amount of unmarried, among 5 biggest :',list(result_sorted.keys())[0])
```

```python
#Show a bar chart of changes in marrital status in Copenhagen from 2008 till now
df = pd.read_csv('https://api.statbank.dk/v1/data/FOLK1A/CSV?delimiter=Semicolon&CIVILSTAND=G&Tid=*&OMR%C3%85DE=084', sep=';')
years = df.loc[:,'TID']
amount_married = df.loc[:,'INDHOLD']

plt.bar(years, amount_married)
plt.title('Changes in married/seperated people in Copenhagen since 2008')
plt.xlabel('Years')
plt.ylabel('Married/Seperated')
plt.xticks(rotation=45, horizontalalignment='right')
ax = plt.gca()
for index, label in enumerate(ax.axes.xaxis.get_ticklabels()):
    if index % 5 != 0:
        label.set_visible(False)
```

```python
import numpy as np
#Show a bar chart of 'Married' and 'Never Married' for all ages in DK (Hint: 2 bars of different color)
df = pd.read_csv('https://api.statbank.dk/v1/data/FOLK1A/CSV?delimiter=Semicolon&ALDER=*&CIVILSTAND=U%2CG', sep=';')

ages = df.ALDER.unique()
ages = list(ages)
ages.pop(0)
unmarried = df.loc[df['CIVILSTAND'] == 'Ugift']['INDHOLD']
unmarried = list(unmarried)
unmarried.pop(0)
married = df.loc[df['CIVILSTAND'] == 'Gift/separeret']['INDHOLD']
married = list(married)
married.pop(0)

x = np.arange(len(ages))
width = 0.35

plt.bar(x - width/2, unmarried, color='b', width=width, label='Unmarried')
plt.bar(x + width/2, married, color='r',  width=width, label='Married')
plt.xticks(rotation=45, horizontalalignment='right')
plt.ylabel('Amount')
plt.title('Married/unmarried for different ages')
plt.xticks(x)
plt.legend()

ax = plt.gca()

ax.set_xticklabels(ages)
for index, label in enumerate(ax.axes.xaxis.get_ticklabels()):
    if index % 10 != 0:
        label.set_visible(False)
```

```python
#Chosen table: "FOLK3"
#Show distribution of peopel with birth day in each month in 2020

df = pd.read_csv('https://api.statbank.dk/v1/data/FOLK3/CSV?delimiter=Semicolon&FMAANED=*', sep=';')
months = df.iloc[1:,0]
birthdays = df.iloc[1:,2]

plt.bar(months, birthdays)
plt.title('Amount of people with birthday in each month in 2020')
plt.xlabel('Months')
plt.ylabel('Amount')
plt.xticks(rotation=45, horizontalalignment='right')
```

```python
#Show which month has the most birthdays each year since 2008
df = pd.read_csv('https://api.statbank.dk/v1/data/FOLK3/CSV?delimiter=Semicolon&Tid=*&FMAANED=001%2C002%2C003%2C004%2C005%2C006%2C007%2C008%2C009%2C010%2C011%2C012', sep=';')
result = {}

months = df.FMAANED.unique()
years = df.TID.unique()

for y in years:
    highest = 0
    month = ''
    for m in months:
        amount = df.loc[df['TID'] == y].loc[df['FMAANED'] == m]['INDHOLD']
        amount = int(amount)
        if amount > highest:
            highest = amount
            month = m
    result[y] = month

print(result)
```

```python
#Show the five years with most births since 1899 (aparently only counts people alive today)
df = pd.read_csv('https://api.statbank.dk/v1/data/FOLK3/CSV?delimiter=Semicolon&FODAAR=*', sep=';')

top_six = df.nlargest(6, 'INDHOLD')
top_five = top_six.drop(0)

years = top_five['FODAAR']
amounts = top_five['INDHOLD']

plt.bar(years, amounts)
plt.title('Top 5 years with most births since 1899')
plt.xlabel('Years')
plt.ylabel('Amount birthed')
    
```

```python
#Find most common date of birth in each month
df = pd.read_csv('https://api.statbank.dk/v1/data/FOLK3/CSV?delimiter=Semicolon&FDAG=*&FMAANED=*', sep=';')

days = df.FDAG.unique()
days = list(days)
days.pop(0)

months = df.FMAANED.unique()
months = list(months)
months.pop(0)

result = {}

for m in months:
    highest_day = ''
    highest_amount = 0
    for d in days:
        amount = df.loc[df['FMAANED'] == m].loc[df['FDAG'] == d]['INDHOLD']
        amount = int(amount)
        if amount > highest_amount:
            highest_amount = amount
            highest_day = d
    result[m] = highest_day

print(result)

#Ved ikke hvorfor den første er den dato med flest fødsler i så mange måneder, så ville lige tjekke på en anden måde.
#Som man kan se her er 9 ud af de 20 dage med flest fødsler i løbet af hele året den første i måneden. 
#Syntes det virker lidt mærkeligt, men kan være det har noget at gøre med hvordan ukendte fødselsdage bliver behandlet?
df = df[df.FDAG != 'I alt']
df = df[df.FMAANED != 'I alt']
print(df.nlargest(20, 'INDHOLD'))
```
